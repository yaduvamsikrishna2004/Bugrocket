```python
import os
import json
import random

DATA_DIR = "./dataset" # Define DATA_DIR
DATA_FILE = os.path.join(DATA_DIR, "coding_train_5000.jsonl") # Define DATA_FILE
os.makedirs(DATA_DIR, exist_ok=True) # Create DATA_DIR if it doesn't exist

NUM_SAMPLES = 10 # Default value, can be changed

# Placeholder for templates. Replace with your actual templates.
templates = [
    ("Write a function to add two numbers.", "def add(a, b):\n    return a + b"),
    ("Explain the concept of recursion.", "Recursion is a programming technique where a function calls itself to solve a problem.")
]

rows = []
while len(rows) < NUM_SAMPLES:
    q, a = random.choice(templates)
    if random.random() < 0.3:
        q = q.replace("Write", random.choice(["Create", "Implement", "Provide"]))
    if random.random() < 0.2:
        a = a + """
# Extra explanation included."""
    rows.append({"instruction": "code", "input": q, "output": a})

with open(DATA_FILE, "w", encoding="utf-8") as f:
    for r in rows:
        f.write(json.dumps(r, ensure_ascii=False) + "\n")

print(f"✔ Dataset generated → {DATA_FILE}")
```
